# Teoria.
* Distribuição Normal(gaussiana) - gauss(média,desvio)
$$f(x) = \frac{e^{\frac{-(x-\mu)^2}{2\sigma^2}}}{\sigma \sqrt{2\pi}}$$

* Distribuição Gamma - gammavariante(alpha,beta)
$$f(x) = \frac{\beta^{\alpha} x^{\alpha -1} e^{-\beta x}}{\Gamma (\alpha) }$$

## Média Aritmética.
$$M = \frac{x_{1} + x_{2} +x_{3} +...+x_{n}}{n}$$
Utilizada para forma mais adequada quando os valores são relativamente uniformes.
## Média Geométrica.
Utilizada mais para volumes de cubos e aplicado mais na financeira que envolve a taxa percentual.
$$G = \sqrt[n]{x_1 * x_2 * ... * x_n}$$
## Média harmônica.
$$H = \frac{n}{1/x_1 + 1/x_2 + 1/x_3 + ... + 1/x_n}$$
A escolha pelo uso da média harmônica para representação da média de um conjunto está ligada a situações que envolvem grandezas inversamente proporcionais, por exemplo a velocidade média, a vazão da água, a densidade, entre outras aplicações na física e na química. Algumas fórmulas da física, na verdade, surgem da média harmônica.
## Mediana.
$$M(x)\begin{cases}
 & \text{ se } x= 2k; x = x_m \\ 
 & \text{ se } x= 2k+1;x = \frac{a+b}{2}
\end{cases}$$
Ela é usada como medida de tendência central, localização quando a distribuição é distorcida, quando os valores dimencionais não são conhecidos.
## Moda.
Moda é o valor que mais aparece dentro de um conjunto quantitativo, com isso para indentificar é nescessario encontrar a frequência de determinados dados.

# Pratica :
* Distribuição uniforme no caso ela nos lembra um retangulo ou seja jogar um dado ou uma moeda etc.
* Distribuição em vermelho é a uniforme.
* Distribuição normal já é são casos que sai da média assim tem como bolsa de valores, problemas que envolvem a natureza etc.
* Distribuição normal é a em verde.
* Distribuição Gamma está em preto.
* Utilizada bastante em servidores colocados em redes ou seja distribuições das filas.

```python
import random as rd
import matplotlib.pyplot as fig



aleat = []
aleat1 = []
for i in range(1000):
    x = rd.gauss(0,3)
    aleat.append(x)
    x = rd.uniform(-1,1)
    aleat1.append(x)
fig.subplot(211)
fig.plot(aleat,'-g')
fig.plot(aleat1,'-r')
#fig.subplot(212)
#fig.hist(aleat,bins=10)
```

```python
aleat = []
for i in range(1000):
    x = rd.gammavariate(1,2)
    aleat.append(x)
fig.subplot(211)
fig.plot(aleat,color='black')
fig.subplot(212)
fig.hist(aleat,bins=20,color='gray')
fig.xlabel('Classes')
fig.ylabel('frequencia')    
```

## Conclusão:
Existe vários tipos de distribuições em que os quais, cada uma tem suas importância no caso a mais comum é a normal, que vem mais voltado para a aleatoriedade da natureza e entre outros como o mercado financeiro no gráfico verde e vermelho podemos notar a grande diferença dos dois.
