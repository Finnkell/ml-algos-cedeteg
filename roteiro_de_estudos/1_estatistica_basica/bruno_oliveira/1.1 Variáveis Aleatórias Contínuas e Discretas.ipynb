### Definição:
Podemos definir as variáveis como qualitativas e quantitativas:
* #### **Qualitativas:** 
São baseadas em qualidades, ou seja, definem uma classificação dos indivíduos. Podem ser nominais ou ordinais:
1. **Nominal:** não possui uma ordem de categoria, por exemplo: cor dos olhos, gênero, bairro, etc.
2. **Ordinal:** possui uma ordem de categorias, por exemplo: colocação (1°, 2°, 3° colocados), patentes do exército (Cabo, Sargento, Capitão).
* #### **Quantitativas:** 
São baseadas em quantidades, ou seja, definem um conjunto de números que fazem sentido. Podem ser contínuas ou discretas:
1. **Discretas:** são definidas por um número finito ou infinito contável de valores, por exemplo: número de filhos, número de funcionários de uma empresa, etc.
2. **Contínuas:** são definidas por uma escala contínua de números na reta real, ou seja, números entre um intervalo ou uma união de números reais, por exemplo: peso, massa, altura, idade, etc.
3. **Mistas:** são variáveis geradas através da junção entre a variável discreta e contínua, ou seja, é uma varável que consegue assumir tanto valores discretos, quanto valores contínuos em um determinado intervalo, por exemplo: lançamento de uma moeda com um jogo de roleta.

### Fórmula:
* $X:\Omega \rightarrow \mathbb{R} \Leftrightarrow \{\omega : X(\omega) \leq x\} \in \mathcal{F}, \forall x \in \mathbb{R}$ 

A função **X** de $\Omega$ em $\mathsf{\mathbb{R}}$ será uma variável aleatória (ou uma função mensurável) se e somente se, para todo valor **$x$**$\in \mathbb{R}$, o conjunto de elementos $\omega$, tais que o valor de função **X** nos elementos **$X(\omega)$** seja menor ou igual a **$x$**, pertencer ao $\sigma-álgebra$ $\mathcal{F}$

### Exemplos c/ Python:
* #### **Discretas:** 
Como um exemplo do uso das variáveis discretas, podemos simular o jogo Jokenpô (pedra, papel e tesoura), pois estamos definindo um número finito de escolhas.

```python
from random import choice, randint
from time import sleep

print('-' * 20)
print('Vamos jogar Jokenpô')
print('-' * 20)

comp = ['PEDRA', 'PAPEL', 'TESOURA']

print('Escolha:') 
print('''
1 - PEDRA
2 - PAPEL
3 - TESOURA
''')

#jog = int(input('Sua opção: '))
jog = randint(1, 3)

print('PENSANDO...')
sleep(1)
ec = choice(comp)

if (ec == 'PEDRA' and jog == 1) or (ec == 'PAPEL' and jog == 2) or (ec == 'TESOURA' and jog == 3):
    print('{}EMPATE{}'.format('\033[1;33m', '\033[m'))
elif (ec == 'PEDRA' and jog == 2) or (ec == 'PAPEL' and jog == 3) or (ec == 'TESOURA' and jog == 1):
    print('{}VC VENCEU! PARABÉNS{}'.format('\033[1;32m', '\033[m'))
elif (ec == 'PEDRA' and jog == 3) or (ec == 'PAPEL' and jog == 1) or (ec == 'TESOURA' and jog == 2):
    print('{}O COMPUTADOR GANHOU{}'.format('\033[1;31m', '\033[m'))
else:
    print('Escolha inválida!')

print('Vc escolheu: {}'.format(comp[jog-1]))
print('Escolha do computador: {}'.format(ec))
```

* #### **Contínuas:**
Como um exemplo do uso das variáveis contínuas, podemos usar um calculadora de índice de massa corporal, pois esse valor varia de pessoa para pessoa, e o resultado está dentro de um intervalo dentro dos números reais.

```python
from random import uniform as uni
import matplotlib.pyplot as plt
import pandas as pd

#tamanho do vetor
tam = 10000
_lista = {'peso':[], 'altura':[], 'imc':[]}
def calc():
    peso = uni(50, 100)
    altura = uni(1.5, 2.10)
    imc = peso / pow(altura, 2)
    _lista['peso'].append(peso)
    _lista['altura'].append(altura)
    
    return imc

def imc(n):
    p = []
    for i in range(n):
        _lista['imc'].append(calc())
    return p

v = []
v = imc(tam)        

df = pd.DataFrame(_lista)
df.plot(x = 'altura', y = 'imc', kind = 'scatter', alpha = .1)
df.plot(x = 'peso', y = 'imc', kind = 'scatter', alpha = .1)
plt.show()
```
