# Principais distribuições: Normal, Bernoulli, Binomial, Uniforme, Poisson, Geométrica

## Teoria

### Distribuição Normal: 
Também chamada de distribuição Gaussiana, é uma distribuição que gera uma curva simétrica, e se essa curva for refinada ela fica em formato de um "sino". A distribuição é feita sobre variáveis contínuas, mas nem todas essas variáveis terão distribuição normal, para que ocorra esta distribuição, a função de densidade de probabilidade uma variável contínua X, deve ser dada pela seguinte fórmula:
 $$f(x) = \frac{1}{\sigma \sqrt{2\pi }}.e^{-\frac{1}{2}.(\frac{x-\mu }{\sigma})^{2}}$$

### Distribuição de Bernoulli: 
É um tipo de distribuição que suporta apenas dois resultados: sucesso ou fracasso. Esses experimentos são chamados ensaios de Bernoulli e originam uma variável aleatória com distribuição de Bernoulli. Essa variável aleatória com probabilidade de sucesso, recebe 1 se o resultado for sucesso e 0 se o resultado for fracasso. A função de probabilidade de X (variável aleatória) é dada pela seguinte função:
$$P(X = x) = p^{x}(1-p)^{(1-x)}$$
Exemplos:
1. Tipos de candles (alta ou baixa)
2. Pratica esportes (sim ou não)
3. Investe em ações (sim ou não)

### Distribuição Binomial: 
É uma distribuição que consiste em repetições independentes de um ensaio de Bernoulli, com a mesma probabilidade de ocorrer um sucesso, ou seja, é uma distribuição onde ocorre n tentativas independentes e repetidas, e cada tentativa só recebe dois valores, 1 ou S para sucesso e 0 ou F para fracasso. É dada pela seguinte função:
$$P(X = x) = \binom{n}{x} p^{x}(1-p)^{n-x}$$

Exemplos:
1. Probabilidade de cair cara em 3 lançamentos de uma moeda
2. Probabilidade de acerto em uma questão de múltipla escolha
3. Qual é a chance de ocorrer três candles de alta consecutivos

### Distribuição Uniforme: 
É a distribuição onde a probabilidade de gerar um ponto em um intervalo contido no espaço amostral, é proporcional ao tamanho do intervalo, ou seja, é definida por um número finito de resultados com chances iguais de acontecer. Como estamos representando possíveis resultados em um intervalo fechado, estamos usando variáveis discretas. A função acumulada dessa distribuição, está representada a seguir:
$$F(x) = P(X\leq x) = \int_{-\infty }^{x}f(x)dx = \left\{\begin{matrix}
0, x < \alpha \\ 
\frac{x-\alpha}{\beta - \alpha}, \alpha\leq x\leq \beta\\ 
1, x \geq \beta
\end{matrix}\right.$$

Exemplos:
1. Probabilidade de peças com defeito em um lote específico de peças
2. Valor resultante do lançamento repetido de um dado com seis lados
3. Probabilidade do índice fechar em alta após um dia de pregão

### Distribuição Poisson: 
É uma distribuição que está associada a eventos raros, ou seja, é uma distribuição aplicada quando o número de ocorrências discretas, é maior que o número médio de ocorrências em um intervalo de tempo.
A função de probabilidade de X (número de ocorrências de um evento no tempo ou no espaço) da distribuição de Poisson, é descrita a seguir:
$$P(X = x) = \frac{e^{-\lambda }\lambda^{x}}{x!}$$

Exemplos:
1. Número de acidentes automotivos em um determinado período
2. Número de ordens enviadas ao mercado por minuto
3. Número de novas pessoas cadastradas em uma corretora, que estão investindo na bolsa

### Distribuição Geométrica: 
É uma distribuição constituída por duas funções de probabilidade discretas:
* O número de ensaios independentes até a ocorrência do primeiro sucesso com probabilidade p
* O número de fracassos antes do primeiro sucesso

A função de probabilidade, é dada por:
$$P(X = x) = p(1-p)^{(x-1)}$$

Exemplos:
1. O número de operações, até que ocorra o primeiro take profit com 100 pontos
2. O número de dias de loss (perca), até que ocorra o primeiro dia de gain (ganho)
3. A quantidade de lançamentos de um dado, até que saia a face com o número 5

## Prática:

### Aplicação:
Temos dois exemplos em python, o primeiro é a demonstração gráfica da distribuição normal ou gaussiana. Já no segundo exemplo, temos outra demonstração por meio de um gráfico, porém temos a exemplificação da distribuição de Poisson 

* **Exemplo 1**

```python
from scipy.stats import norm
import numpy as np
import matplotlib.pyplot as plt

x_rvs = norm.rvs(loc=0, scale=1, size=100000, random_state=3)

mu, sigma = norm.fit(x_rvs)
x = np.linspace(-5, 5, 5000)
fig, ax = plt.subplots(figsize=(16, 6))
ax.hist(x_rvs, bins=1000, density=True)
ax.plot(x, norm.pdf(x, loc=mu, scale=sigma), lw=5)
plt.show()
```

* **Exemplo 2**

```python
from scipy.stats import poisson
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

x_rvs = pd.Series(poisson.rvs(1.2, size=100000, random_state=2))
data = x_rvs.value_counts().sort_index().to_dict()

fig, ax = plt.subplots(figsize=(16, 6))
ax.bar(range(len(data)), list(data.values()), align='center')
plt.xticks(range(len(data)), list(data.keys()))
plt.show()
```
